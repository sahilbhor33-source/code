**Step 1: Import TensorFlow**

```python
import tensorflow as tf
print("TensorFlow Version:", tf.__version__)
```

**Step 2: Creating Tensors**

```python
# Scalar
a = tf.constant(10)

# Vector
b = tf.constant([1,2,3])

# Matrix
c = tf.constant([[1,2],[3,4]])

print("Scalar:", a)
print("Vector:", b)
print("Matrix:\n", c)
```

**Step 3: Basic Mathematical Operations**

```python
x = tf.constant(20)
y = tf.constant(5)

print("Addition:", tf.add(x, y))
print("Subtraction:", tf.subtract(x, y))
print("Multiplication:", tf.multiply(x, y))
print("Division:", tf.divide(x, y))
```

**Step 4: TensorFlow Variables**

```python
W = tf.Variable(5.0)
b = tf.Variable(2.0)

print("Weight:", W)
print("Bias:", b)
```

**Step 5: Simple Linear Model**

```python
# Input
x = tf.constant(3.0)

# Linear equation y = Wx + b
y = W * x + b

print("Output:", y)
```

**Step 6: Simple Neural Network Using Keras**

```python
from tensorflow import keras
import numpy as np

model = keras.Sequential([
    keras.layers.Dense(1, input_shape=[1])
])

model.compile(optimizer='sgd', loss='mean_squared_error')

X = np.array([1,2,3,4], dtype=float)
Y = np.array([2,4,6,8], dtype=float)

model.fit(X, Y, epochs=100)

# Correct Prediction
prediction = model.predict(np.array([[5.0]]))
print("Prediction for 5:", prediction)
```

**Using CNN **

```python
# ==========================================
# CNN USING TENSORFLOW - INBUILT DATASET
# ==========================================

# Step 1: Import Libraries
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt
import numpy as np

print("TensorFlow Version:", tf.__version__)

# Step 2: Load Inbuilt MNIST Dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

print("Training Data Shape:", x_train.shape)

# Step 3: Normalize Data
x_train = x_train / 255.0
x_test = x_test / 255.0

# Step 4: Reshape for CNN (Add Channel Dimension)
x_train = x_train.reshape(-1,28,28,1)
x_test = x_test.reshape(-1,28,28,1)

# Step 5: Build CNN Model
model = keras.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# Step 6: Compile Model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Step 7: Train Model
history = model.fit(x_train, y_train, epochs=5, validation_data=(x_test, y_test))

# Step 8: Evaluate Model
test_loss, test_accuracy = model.evaluate(x_test, y_test)
print("Test Accuracy:", test_accuracy)

# Step 9: Plot Accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title("Accuracy Graph")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend(["Train", "Validation"])
plt.show()

# Step 10: Prediction Example
predictions = model.predict(x_test)

plt.imshow(x_test[0].reshape(28,28), cmap='gray')
plt.title("Predicted: " + str(np.argmax(predictions[0])) +
          " | Actual: " + str(y_test[0]))
plt.axis('off')
plt.show()
```
